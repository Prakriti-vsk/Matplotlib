```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.step(x,y)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,3,4,2,5]

plt.step(x,y)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]


plt.grid()
plt.step(x,y)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.title('python')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid()
plt.step(x,y)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]
plt.step(x,y,color='r',marker='o')


plt.title('python')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid()

plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]
plt.step(x,y,color='r',marker='o',ms=10)


plt.title('python')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid()

plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]
plt.step(x,y,color='r',marker='o',ms=10,mfc='g')


plt.title('python')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid()

plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]
plt.step(x,y,color='r',marker='o',ms=10,mfc='g',label='python')


plt.title('python')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.grid()

plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]
plt.step(x,y,color='r',marker='o',ms=10,mfc='g',label='python')


plt.title('python')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend(loc=4)
plt.grid()

plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]
plt.step(x,y,color='g',marker='o',ms=10,mfc='g',label='python')


plt.title('python')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.grid()

plt.show()
```
