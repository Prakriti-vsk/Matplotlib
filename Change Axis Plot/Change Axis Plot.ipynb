```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Axis Plot

```python
x=[1,2,3,4,5]
y=[3,1,5,2,4]

plt.plot(x,y,color='r')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[3,1,5,2,4]

plt.plot(x,y,color='r')

plt.xticks(x)
plt.yticks(x)
plt.show()
```

```python
x=[1,2,3,4]
y=[3,1,5,2]

plt.plot(x,y,color='r')

plt.xticks(x,labels=['python','c','c++','java'])
plt.yticks(x,labels=['python','c','c++','java'])
plt.show()
```

```python
x=[1,2,3,4]
y=[3,1,5,2]

plt.plot(x,y,color='r')

plt.xlim(0,10)
plt.ylim(0,10)

plt.show()
```

```python
x=[1,2,3,4]
y=[3,1,5,2]

plt.plot(x,y,color='r')

#plt.xlim(0,10)
#plt.ylim(0,10)

plt.axis([0,10,0,7])

plt.show()
```
