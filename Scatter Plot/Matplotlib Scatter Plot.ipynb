# Scatter Plot

```python
x=[1,2,3,4,5,6,7]
y=[2,3,1,4,5,3,6]
plt.scatter(x,y)
plt.show()
```

```python
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
plt.scatter(day,no,color='r')
plt.title('Scatter plot',fontsize=12)
plt.xlabel('Day',fontsize=12)
plt.ylabel('No',fontsize=12)
plt.show()
```

```python
import matplotlib.pyplot as plt

day = [1, 2, 3, 4, 5, 6, 7]
no = [2, 3, 1, 4, 5, 3, 6]

colors = ['red', 'blue', 'green', 'yellow', 'red', 'blue', 'violet']

plt.scatter(day, no, c=colors)
plt.title('Scatter plot', fontsize=12)
plt.xlabel('Day', fontsize=12)
plt.ylabel('No', fontsize=12)
plt.show()
```

```python
import matplotlib.pyplot as plt

day = [1, 2, 3, 4, 5, 6, 7]
no = [2, 3, 1, 4, 5, 3, 6]

colors = ['red', 'blue', 'green', 'yellow', 'red', 'blue', 'violet']
sizes=[100,200,400,500,200,100,700]

plt.scatter(day, no, c=colors,s=sizes)
plt.title('Scatter plot', fontsize=12)
plt.xlabel('Day', fontsize=12)
plt.ylabel('No', fontsize=12)
plt.show()
```

```python
import matplotlib.pyplot as plt

day = [1, 2, 3, 4, 5, 6, 7]
no = [2, 3, 1, 4, 5, 3, 6]

colors = [10,45,39,56,12,45,15]
sizes=[100,200,400,500,200,100,700]

plt.scatter(day, no, c=colors,s=sizes,cmap='BrBG',alpha=0.74)
t=plt.colorbar()
t.set_label('ColorBar', fontsize=12)

plt.title('Scatter plot', fontsize=12)
plt.xlabel('Day', fontsize=12)
plt.ylabel('No', fontsize=12)
plt.show()
```

```python
import matplotlib.pyplot as plt

day = [1, 2, 3, 4, 5, 6, 7]
no = [2, 3, 1, 4, 5, 3, 6]
no2=[3,2,4,5,1,6,2]
colors = [10,45,39,56,12,45,15]
sizes=[100,200,400,500,200,100,700]

plt.scatter(day, no, c=colors,s=sizes,cmap='BrBG',alpha=0.74)
plt.scatter(day, no2, c='r',s=sizes,alpha=0.74)
t=plt.colorbar()
t.set_label('ColorBar', fontsize=12)

plt.title('Scatter plot', fontsize=12)
plt.xlabel('Day', fontsize=12)
plt.ylabel('No', fontsize=12)
plt.show()
```
