# Text Plot

```python
x=[1,2,3,4,5]
y=[3,1,5,2,6]

plt.plot(x,y)

plt.show()
```

```python
x=[1,2,3,4,5]
y=[3,1,5,2,6]

plt.title('python')

plt.plot(x,y)

plt.show()
```

```python
x=[1,2,3,4,5]
y=[3,1,5,2,6]

plt.title('python',fontsize=20)

plt.plot(x,y)

plt.show()
```

```python
x=[1,2,3,4,5]
y=[3,1,5,2,6]

plt.title('python',fontsize=12)
plt.xlabel('days',fontsize=12)
plt.ylabel('week',fontsize=12)

plt.plot(x,y)

plt.show()
```

```python
x=[1,2,3,4,5]
y=[3,1,5,2,6]

plt.title('python',fontsize=12)
plt.xlabel('days',fontsize=12)
plt.ylabel('week',fontsize=12)

plt.plot(x,y)

plt.text(2,5,'Python')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[3,1,5,2,6]

plt.title('python',fontsize=12)
plt.xlabel('days',fontsize=12)
plt.ylabel('week',fontsize=12)

plt.plot(x,y)

plt.text(2,5,'Java',fontsize=12,style='italic')
plt.show()
```
