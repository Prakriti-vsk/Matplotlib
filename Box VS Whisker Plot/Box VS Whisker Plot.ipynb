```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Box VS Whisker Plot

```python
x=[10,20,30,40,50,60,70]

plt.boxplot(x)
plt.show()
```

```python
x=[10,20,30,40,50,60,70,120]

plt.boxplot(x)
plt.show()
```

```python
x=[10,20,30,40,50,60,70]

plt.boxplot(x,notch=True)
plt.show()
```

```python
x=[10,20,30,40,50,60,70]

plt.boxplot(x,notch=False)
plt.show()
```

```python
x=[10,20,30,40,50,60,70]

plt.boxplot(x,vert=False)
plt.show()
```

```python
x=[10,20,30,40,50,60,70]

plt.boxplot(x,widths=0.4)
plt.show()
```

```python
x=[10,20,30,40,50,60,70]

plt.boxplot(x,labels=['python'])
plt.show()
```

```python
x=[10,20,30,40,50,60,70]

plt.boxplot(x,labels=['python'],patch_artist=True)
plt.show()
```

```python
x=[10,20,30,40,50,60,70]

plt.boxplot(x,labels=['python'],patch_artist=True,showmeans=True)
plt.show()
```

```python
x=[10,20,30,40,50,60,70]

plt.boxplot(x,labels=['python'],showmeans=True)
plt.show()
```

```python
x=[10,20,30,40,50,60,70,130]

plt.boxplot(x,labels=['python'],showmeans=True)
plt.show()
```

```python
x=[10,20,30,40,50,60,70,120]

plt.boxplot(x,labels=['python'],showmeans=True,whis=2)
plt.show()
```

```python
x=[10,20,30,40,50,60,70]

plt.boxplot(x,labels=['python'],
            showmeans=True,sym="g+")
plt.show()
```

```python
x=[10,20,30,40,50,60,70]

plt.boxplot(x,labels=['python'],
            showmeans=True,sym="g+",boxprops=dict
           (color='r'),capprops=dict(color='g'))
plt.show()
```

```python
x=[10,20,30,40,50,60,70]

plt.boxplot(x,labels=['python'],
            showmeans=True,sym="g+",boxprops=dict
           (color='r'),capprops=dict(color='g'),whiskerprops=dict(color='y'))
plt.show()
```

```python
x=[10,20,30,40,50,60,70,120]

plt.boxplot(x,labels=['python'],
            showmeans=True,sym="g+",boxprops=dict
           (color='r'),capprops=dict(color='g'),whiskerprops=dict(color='y'),flierprops=dict(markeredgecolor='b'))
plt.show()
```

```python
x=[10,20,30,40,50,60,70,120]
x1=[10,20,40,50,30,60,90]
plt.boxplot(x,labels=['python'],
            showmeans=True,sym="g+",boxprops=dict
           (color='r'),capprops=dict(color='g'),whiskerprops=dict(color='y'),flierprops=dict(markeredgecolor='b'))

plt.boxplot(x1,labels=['python'],
            showmeans=True,sym="g+",boxprops=dict
           (color='r'),capprops=dict(color='g'),whiskerprops=dict(color='y'),flierprops=dict(markeredgecolor='b'))

plt.show()
```
