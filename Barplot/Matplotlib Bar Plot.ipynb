# Bar Plot

```python
x=['python','c','c++','java']
y=[85,70,60,82]
plt.bar(x,y,color=c,width=0.5,edge=0.3)
c=('r','y','b','g')
plt.xlabel('Language',fontsize=12)
plt.ylabel('No',fontsize=12)
plt.title('WorkfromHome',fontsize=12)
plt.show()
```

```python
x=['python','c','c++','java']
y=[85,70,60,82]
plt.bar(x,y,color=c,width=0.5,edgecolor='grey',label='popularity')
c=('r','grey','b','g')
plt.xlabel('Language',fontsize=12)
plt.ylabel('No',fontsize=12)
plt.title('WorkfromHome',fontsize=12)
plt.legend()
plt.show()
```

```python
x=['python','c','c++','java']
y=[85,70,60,82]
z=[20,30,40,50]
plt.bar(x,y,color='yellow',width=0.5,edgecolor='grey',label='popularity')
plt.bar(x,z,color='grey',width=0.5,edgecolor='grey',label='popularity1')
c=('r','s','b','g')
plt.xlabel('Language',fontsize=12)
plt.ylabel('No',fontsize=12)
plt.title('WorkfromHome',fontsize=12)
plt.legend()
plt.show()
```

```python
x=['python','c','c++','java']
y=[85,70,60,82]
z=[20,30,40,50]
p=np.arange(len(x))
plt.xlabel('Language',fontsize=12)
plt.ylabel('No',fontsize=12)
plt.title('WorkfromHome',fontsize=12)
width=0.2
plt.bar(x,y,width,color='orange',edgecolor='grey',label='popularity')
plt.bar(x,z,width,color='grey',edgecolor='grey',label='popularity1')
plt.legend()
plt.show()
```

```python
x=['python','c','c++','java']
y=[85,70,60,82]
z=[20,30,40,50] 

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel('Language',fontsize=12)
plt.ylabel('No',fontsize=12)
plt.title('WorkfromHome',fontsize=12)

plt.bar(p,y,width,color='grey',edgecolor='grey',label='popularity')
plt.bar(p1,z,width,color='black',edgecolor='grey',label='popularity1')

plt.xticks(p+width/2,x)
plt.legend()
plt.show()
```

```python
x=['python','c','c++','java']
y=[85,70,60,82]
z=[20,30,40,50] 

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel('Language',fontsize=12)
plt.ylabel('No',fontsize=12)
plt.title('WorkfromHome',fontsize=12)

plt.bar(p,y,width,color='grey',edgecolor='grey',label='popularity')
plt.bar(p1,z,width,color='black',edgecolor='grey',label='popularity1')

plt.xticks(p+width/2,x,rotation=45)
plt.legend()
plt.show()
```

```python
x=['python','c','c++','java']
y=[85,70,60,82]
z=[20,30,40,50] 

plt.xlabel('Language',fontsize=12)
plt.ylabel('No',fontsize=12)
plt.title('WorkfromHome',fontsize=12)

plt.barh(p,y,color='grey',edgecolor='grey',label='popularity')
plt.barh(p1,z,color='black',edgecolor='grey',label='popularity1')


plt.legend()
plt.show()
```
