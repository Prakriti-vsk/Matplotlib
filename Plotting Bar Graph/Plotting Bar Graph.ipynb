```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Plotting Bar Plot

```python
x=['Python','C','C++','Java']
y=[85,70,60,62]
plt.xlabel('Language',fontsize=12)
plt.ylabel('No.',fontsize=12)
plt.title('Jaadu',fontsize=12)
colors = ['red', 'blue', 'green', 'yellow', 'red', 'blue', 'violet']
plt.bar(x,y,width=0.4,color='r')
plt.show()
```

```python
x=['Python','C','C++','Java']
y=[85,70,60,62]
z=[20,30,40,50]
plt.xlabel('Language',fontsize=12)
plt.ylabel('No.',fontsize=12)
plt.title('Jaadu',fontsize=12)
colors = ['red', 'blue', 'green', 'orange']
plt.bar(x,y,width=0.4,color=colors,align='center',edgecolor='g',label='popularity1')
plt.bar(x,z,width=0.4,color='grey',align='center',label='popularity')
plt.legend()
plt.show()
```

```python
x=['Python','C','C++','Java']
y=[85,70,60,62]
z=[20,30,40,50]
p=np.arange(len(x))
width=0.2

p1=[j+width for j in p]
plt.xlabel('Language',fontsize=12)
plt.ylabel('No.',fontsize=12)
plt.title('Jaadu',fontsize=12)

plt.bar(p,y,width,color='white',align='center',edgecolor='g',label='popularity1')
plt.bar(p1,z,width,color='grey',align='center',label='popularity')
plt.legend()
plt.show()
```

```python
x=['Python','C','C++','Java']
y=[85,70,60,62]
z=[20,30,40,50]
p=np.arange(len(x))
width=0.2

p1=[j+width for j in p]
plt.xlabel('Language',fontsize=12)
plt.ylabel('No.',fontsize=12)
plt.title('Jaadu',fontsize=12)

plt.bar(p,y,width,color='white',align='center',edgecolor='g',label='popularity1')
plt.bar(p1,z,width,color='grey',align='center',label='popularity')

plt.xticks(p+width/2,x,rotation=45)
plt.legend()
plt.show()
```

```python
x=['Python','C','C++','Java']
y=[85,70,60,62]
z=[20,30,40,50]
p=np.arange(len(x))
width=0.2

p1=[j+width for j in p]
plt.xlabel('Language',fontsize=12)
plt.ylabel('No.',fontsize=12)
plt.title('Jaadu',fontsize=12)

plt.barh(p,y,width,color='white',align='center',edgecolor='g',label='popularity1')
plt.barh(p1,z,width,color='white',align='center',edgecolor='r',label='popularity')

plt.xticks(p+width/2,x,rotation=45)
plt.legend()
plt.show()
```
