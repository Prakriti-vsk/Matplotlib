```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Stem Plot

```python
x=[1,2,3,4,5,6]
y=[2,2,5,6,4,3]

plt.stem(x,y)
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,2,5,6,4,3]

plt.stem(x,y,linefmt=":")
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,2,5,6,4,3]

plt.stem(x,y,linefmt=":",markerfmt="r+")
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,2,5,6,4,3]

plt.stem(x,y,linefmt=":",markerfmt='ro')
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,2,5,6,4,3]

plt.stem(x,y,linefmt=":",markerfmt='ro',bottom=1)
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,2,5,6,4,3]

plt.stem(x,y,linefmt=":",markerfmt='ro',bottom=9)
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,2,5,6,4,3]

plt.stem(x,y,linefmt=":",markerfmt='ro',bottom=9,basefmt='g')
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,2,5,6,4,3]

plt.stem(x,y,linefmt=":",markerfmt='ro',bottom=9,basefmt='g',label='python')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6]
y = [2, 2, 5, 6, 4, 3]

plt.stem(x,y,linefmt=":", markerfmt='ro',bottom=0,basefmt='g', label='python',use_line_collection=True,orientation="vertical")
plt.legend()
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,2,5,6,4,3]

plt.stem(x,y,linefmt=":",markerfmt='ro',bottom=9,basefmt='g',label="python",orientation='horizontal')
plt.show()
```
