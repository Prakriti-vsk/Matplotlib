```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Stack Plot

```python
x=[1,2,3,4,5]
area=[1,2,3,4,5]
plt.stackplot(x,area)
plt.show()
```

```python
x=[1,2,3,4,5]
area=[2,5,4,1,2]
plt.stackplot(x,area)
plt.show()
```

```python
x=[1,2,3,4,5]
area1=[2,3,2,4,5]
area2=[1,3,2,4,2]
plt.stackplot(x,area1,area2)
plt.show()
```

```python
x=[1,2,3,4,5]
area1=[2,3,2,4,5]
area2=[1,3,2,4,2]
area3=[1,2,7,3,5]
plt.stackplot(x,area1,area2,area3)
plt.show()
```

```python
x=[1,2,3,4,5]
area1=[2,3,2,4,5]
area2=[1,3,2,4,2]
area3=[1,2,7,3,5]

l=['area1','area2','area3']
plt.stackplot(x,area1,area2,area3,labels=l)
plt.legend(loc=2)
plt.show()
```

```python
x=[1,2,3,4,5]
area1=[2,3,2,4,5]
area2=[1,3,2,4,2]
area3=[1,2,7,3,5]

l=['area1','area2','area3']
plt.stackplot(x,area1,area2,area3,labels=l,colors=['r','y','g','b'])
plt.legend(loc=2)
plt.show()
```

```python
x=[1,2,3,4,5]
area1=[2,3,2,4,5]
area2=[1,3,2,4,2]
area3=[1,2,7,3,5]

l=['area1','area2','area3']
plt.stackplot(x,area1,area2,area3,labels=l,colors=['r','y','g','b'],baseline='wiggle')
plt.legend(loc=2)
plt.show()
```

```python
x=[1,2,3,4,5]
area1=[2,3,2,4,5]
area2=[1,3,2,4,2]
area3=[1,2,7,3,5]

l=['area1','area2','area3']
plt.stackplot(x,area1,area2,area3,labels=l,colors=['r','y','g','b'],baseline='wiggle')

plt.title('python')
plt.xlabel('X-axis')
plt.ylabel('y=axis')
plt.legend(loc=2)
plt.show()
```

```python
x=[1,2,3,4,5]
area1=[2,3,2,4,5]
area2=[1,3,2,4,2]
area3=[1,2,7,3,5]

l=['area1','area2','area3']
plt.stackplot(x,area1,area2,area3,labels=l,colors=['r','y','g','b'],baseline='wiggle')

plt.title('python')
plt.grid(color='r')
plt.xlabel('X-axis')
plt.ylabel('y=axis')
plt.legend(loc=2)
plt.show()
```
